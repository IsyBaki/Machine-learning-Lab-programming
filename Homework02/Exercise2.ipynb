```python
#Linear regression

"""
 (a) Exploratory Analysis: Load the dataset insurance.csv and perform basic exploratory data analysis
(EDA).
• Display the first few rows and dataset information (e.g., column names, data types, and missing
values).
• Plot basic distributions (e.g., histograms or boxplots) of numerical features such as age, bmi, and
charges.
• Summarize the statistics of the dataset using describe()

"""
import pandas as pd
import matplotlib.pyplot as plt


# read the data 
data = pd.read_csv('insurance.csv')
display(data.head())


print(data.info())
# missing data
print(data.isnull().sum())

#summary statistics
print(data.describe())
```

```python
#Histograms

numerical_features = ['age', 'bmi', 'charges']
data[numerical_features].hist(figsize=(10, 5), bins=20)
plt.suptitle("Distribution of Numerical Features")
plt.show
```

```python
# Histogram  age
plt.hist(data['age'], bins=20, color='blue', edgecolor='black')
plt.title("Distribution of the Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# Histogram sex
plt.hist(data["sex"], bins=20, color="orange", edgecolor="black")
plt.title("Distribution of the Sex")
plt.xlabel("Sex")
plt.ylabel("Count")
plt.show()
```

```python
#Boxplots for outlier detection
plt.figure(figsize=(10, 5))
for i, col in enumerate(numerical_features):
    plt.subplot(1, 3, i + 1)
    plt.boxplot(data[col])
    plt.title(col)
plt.tight_layout()
plt.show()
```
